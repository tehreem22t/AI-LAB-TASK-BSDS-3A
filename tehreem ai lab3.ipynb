```python
class ModelBasedReflexAgent:
    def __init__(self):
        self.last_action = "OFF"  # Heater starts OFF

    def decide(self, temperature):
        if temperature < 20 and self.last_action != "ON":
            self.last_action = "ON"
            return "Turning heater ON"
        elif temperature > 25 and self.last_action != "OFF":
            self.last_action = "OFF"
            return "Turning heater OFF"
        else:
            return f"Keeping heater {self.last_action}"

agent = ModelBasedReflexAgent()

temperatures = [18, 19, 21, 23, 26, 24, 19, 22]

for temp in temperatures:
    action = agent.decide(temp)
    print(f"Temp: {temp}°C -> {action}")
```
