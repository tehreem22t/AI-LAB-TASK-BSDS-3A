```python
def luhn_check(card_number):
    digits = [int(d) for d in card_number if d.isdigit()]
    check_sum = 0
    is_second = False

    for i in range(len(digits) - 1, -1, -1):
        d = digits[i]
        if is_second:
            d = d * 2
            if d > 9:
                d -= 9
        check_sum += d
        is_second = not is_second

    return check_sum % 10 == 0

card = "4539578763621486"
if luhn_check(card):
    print("Valid card number")
else:
    print("Invalid card number")
```

```python
def remove_punctuation(text):
    punctuations = '''!()-[]{};:'",<>./?@#$%^&*_~'''
    result = ""
    for char in text:
        if char not in punctuations:
            result += char
    return result

sentence = "Hello, world! It's a beautiful day :)"
cleaned = remove_punctuation(sentence)
print("Without punctuation:", cleaned)
```

```python
def sort_sentence(sentence):
    words = sentence.lower().split()

    for i in range(len(words)):
        for j in range(i + 1, len(words)):
            if words[i] > words[j]:
                words[i], words[j] = words[j], words[i]

    return " ".join(words)

sentence = "Banana apple Mango cherry"
sorted_sentence = sort_sentence(sentence)
print("Sorted sentence:", sorted_sentence)
```
